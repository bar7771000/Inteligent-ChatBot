##### Let's import packages we're going to use


```python
import random
import json
import pickle
import numpy as np
import nltk
from nltk.stem import WordNetLemmatizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD
import os
```

#### Were going to need:
* JSON package to read a file in .json format to access training data
* Pickle to save the words and classes list to a compressed file
* Word lemmatizer to transform words to a simplest form
* Keras to build our neural network and train our model


### Next on our list, we have to initialize objects and set variables

```python
lemmatizer = WordNetLemmatizer()
words = []  #tokenized list of words
classes = []  #list of tags
documents = [] #nested list with tokenized words and assigned tags
ignore_letters = ['?', '!', "'", ',', '.']  #array of an exempted words
```

### Before we begin the classification model training, we have to build vocabulary first. Patterns are processed to create a dictionary and each word is lemmetized to produce a generic root, a base form of a given expression.

```python
intents = json.loads(open("intents.json").read())

for intent in intents["intents"]: #looping through intents json
    for pattern in intent["patterns"]: #nested loop through the patterns in insents
        word_list = nltk.word_tokenize(pattern) #splits the sentence in a separate elements of a list
        words.extend(word_list) #appending content to the list
        documents.append((word_list, intent["tag"]))
        if intent["tag"] not in classes:
            classes.append(intent["tag"])
            
words = [lemmatizer.lemmatize(word) for word in words if word not in ignore_letters]
```

### Once we have access to our data, we can sort it and eliminate duplicates with the sorted metthod and save our data to a pickle file

```python

words = sorted(set(words)) #eliminates duplicates
classes = sorted(set(classes))

pickle.dump(words, open("words.pkl", "wb")) #save word list to a file
pickle.dump(classes, open("classes.pkl", "wb"))
```

### Let's prepare training data for our model where X are patterns and Y's are intents

```python
training = []
output_empty = [0] * len(classes)

for document in documents: #creating different combination
    bag = []
    word_patterns = document[0]
    word_patterns = [lemmatizer.lemmatize(word.lower()) for word in word_patterns]
    for word in words:
        bag.append(1) if word in word_patterns else bag.append(0)

    output_row = list(output_empty)
    output_row[classes.index(document[1])] = 1
    training.append([bag, output_row])


random.shuffle(training)
training = np.array(training, dtype=object)

train_x = list(training[:, 0])
train_y = list(training[:, 1])
```

### Time to build the model!

```python
#Building the layers
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation="relu"))
model.add(Dropout(0.4))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.4))
model.add(Dense(len(train_y[0]), activation="softmax"))

model.summary()
```

        We will be useing simple Keras Sequential model consist of three layers. Classification output will be multiclass array, which would help to recognize encoded intent using softmax activation function. With the model summary we can see the overall structure ofthe model and how many parameters does each layer take as an input. As we can see, my model includes 2 Dropouts with value of 0.4 to prevent model from overfitting by randomly setting the inputs units to 0 with a given rate.

### Once we have our model structure, we can start training it

```python
sgd = SGD(learning_rate=0.01, decay=1e-6, momentum=0.9, nesterov=True) #Setting custom optimizer

#Compiling model with Nesterov accelerated gradient to achieve best results
model.compile(loss="categorical_crossentropy", optimizer=sgd, metrics=["accuracy"])

#Training the model
trained_model = model.fit(np.array(train_x), np.array(train_y), epochs=250, batch_size=5, verbose=1)
model.save("inteligent_chatbot_model.model", trained_model) #Saving the trained model

print("Finished!")

```

### Now we should have a saved file in directory folder with our trained model.

### Continuing  with the project, before we make the predictions we need to write couple of function which will help us to clean the user input and transform it into the format that will be more readable to our model

```python
def organize_sentence(sentence):  #pre-processing the given sentence
    sentence_words = nltk.word_tokenize(sentence)
    sentence_words = [lemmatizer.lemmatize(word) for word in sentence_words]
    return sentence_words

def list_of_words(sentence): #checking for the word presence(1: present, 0:absent)
    sentence_words = organize_sentence(sentence)
    word_list = [0] *len(words)
    for w in sentence_words:
        for i, word in enumerate(words):
            if word == w:
                word_list[i] = 1
    return np.array(word_list)

def predict_class(sentence): 
    return_list = []
    bow = list_of_words(sentence)
    result = model.predict(np.array([bow]))[0]
    ERROR_TRESHOLD = 0.25
    # filtering predictions below a threshold, and returning intent index
    results = [[i,r] for i,r in enumerate(result) if r > ERROR_TRESHOLD]
    #sorting in descending order to get the highrst result first
    results.sort(key=lambda x:x[1], reverse=True) 

    for r in results:
        return_list.append({"intent" : classes[r[0]], "probability" : str(r[1])})
    return return_list

def get_response(intents_list, intents_json):
    tag = intents_list[0]["intent"]
    list_of_intents = intents_json["intents"]
    for i in list_of_intents:
        if i["tag"] == tag:
            result = random.choice(i["responses"])
            break
    return result
```

### And to finish it, we have to make the model interactive with user. For that, we can use the loop taking user input and to processing it with our functions and to predict the outcome as shown below: 

```python
print("ChatBot Is Running!")

while True:
    message = input("")
    ints = predict_class(message)
    res = get_response(ints, intents)
    print(res)
```
